```python
n_obj_seq = 12
n_mthd_seq = 11
n_deal_seq = 9
```

## Pad and convert to array

```python
def pad(encoded_text: List[int], n_seq: int) -> np.ndarray:
  len_text = len(encoded_text)
  if len_text >= n_seq:
    return np.array(encoded_text[:n_seq])
  else:
    return np.pad(encoded_text, (0, n_seq - len_text))
```

```python
def concat_padded_texts(row: pd.DataFrame) -> np.ndarray:
  return np.concatenate([row.text_obj, row.text_mthd, row.text_deal])
```

### train set

```python
with open('data/encoded_trainset.pkl', 'rb') as f:
  df_padded_train = pickle.load(f)
```

```python
df_padded_train
```

```python
df_padded_train['text_obj'] = df_padded_train.text_obj.apply(lambda x: pad(x, n_obj_seq))
df_padded_train['text_mthd'] = df_padded_train.text_mthd.apply(lambda x: pad(x, n_mthd_seq))
df_padded_train['text_deal'] = df_padded_train.text_deal.apply(lambda x: pad(x, n_deal_seq))
```

```python
df_padded_train['text_padded'] = df_padded_train.apply(concat_padded_texts, axis=1)
```

```python
padded_train_array = np.stack(df_padded_train.text_padded)
```

```python
padded_train_array.shape
```
