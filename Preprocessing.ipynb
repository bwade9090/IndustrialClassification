# Split train set and validation set

```python
from sklearn.model_selection import train_test_split
```

```python
df.dtypes
```

```python
df.digit_3.value_counts()
```

```python
label_3_list = sorted(list(df.digit_3.unique()))
```

```python
df_sub_train_list = []
df_sub_test_list = []
```

```python
for label in label_3_list:
    df_sliced = df[df.digit_3 == label]
    
    if len(df_sliced) == 1:
        df_sub_train_list.append(df_sliced)
        
    else:
        df_sub_train, df_sub_test = train_test_split(df_sliced, train_size=0.7, test_size=0.3, random_state=0)
        df_sub_train_list.append(df_sub_train)
        df_sub_test_list.append(df_sub_test)
```

```python
len(df_sub_train_list), len(df_sub_test_list)
```

```python
df_train = pd.concat(df_sub_train_list)
df_test = pd.concat(df_sub_test_list)
```

```python
df_train = df_train.sort_index()
```

```python
df_test = df_test.sort_index()
```

```python
df_train.to_csv('data/trainset.csv')
df_test.to_csv('data/validationset.csv')
```
